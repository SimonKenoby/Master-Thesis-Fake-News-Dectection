```python
from pymongo import MongoClient
from bson.objectid import ObjectId
import altair as alt
import pandas as pd
import math
```

```python
client = MongoClient('192.168.178.25', 27017)
db = client.TFE
collection = db.results
```

```python
results = pd.DataFrame(columns = ['type', 'recall', 'precision', 'C'])
```

```python
db = collection.find_one({'_id' : ObjectId("5cd54bdd1459e7270e6a74ad")})
```

```python
for report in db['report']:
    clr = report['classification_report']
    results = results.append({'type' : 'fake', 'recall' : clr['fake']['recall'], 'precision' : clr['fake']['precision'], 'C' : report['C']}, ignore_index=True)
    results = results.append({'type' : 'reliable', 'recall' : clr['reliable']['recall'], 'precision' : clr['reliable']['precision'], 'C' : report['C']}, ignore_index=True)
```

```python
results[results['type'] == 'fake']
```

```python
line = alt.Chart(results).mark_line().encode(
    alt.X('C'),
    alt.Y('recall',
        scale=alt.Scale(zero=False)),
    color='type:N'
)


dots = alt.Chart(results).mark_point().encode(
    alt.X('C'),
    alt.Y('recall',
        scale=alt.Scale(zero=False)),
    color='type:N'
)

chart = alt.layer(
    line, dots
).configure_legend(
    titleAlign='left'
)

chart.save('output/LinearSVC_recall.png', webdriver='firefox')
```

```python
line = alt.Chart(results).mark_line().encode(
    alt.X('C'),
    alt.Y('precision',
        scale=alt.Scale(zero=False)),
    color='type:N'
)


dots = alt.Chart(results).mark_point().encode(
    alt.X('C'),
    alt.Y('precision',
        scale=alt.Scale(zero=False)),
    color='type:N'
)

chart = alt.layer(
    line, dots
).configure_legend(
    titleAlign='left'
)

chart.save('output/LinearSVC_precision.png', webdriver='firefox')
```
